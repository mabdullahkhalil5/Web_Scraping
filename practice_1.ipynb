```python
# import libraries
import requests 
from bs4 import BeautifulSoup
```

```python
# request for scrap
head={
    "User-Agent":"Mozilla/5.0"
}
response = requests.get("https://www.cisco.com/",headers=head)
```

```python
# get title of cisco webpage
soup = BeautifulSoup(response.text,"html.parser")
print("Title: ",soup.title.text)
```

```python
# get links on cisco webpage
links = soup.find_all("a")
for link in links:
    print(link.get("href"))
```

```python
# get all h2 headings
heading = soup.find_all("h2")
for h in heading:
    print(h.text)
```
